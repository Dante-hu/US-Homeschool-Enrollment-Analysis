# Data Cleaning
Data cleaning is a crucial phase in our project that involves refining and organizing the dataset to ensure its quality and integrity for subsequent analysis. During this process, we will address various aspects, including handling missing values and ensuring consistency in data types. By employing these data manipulation techniques, we aim to create a clean and reliable dataset that aligns with the project's focus. ## Step 1
Let's clean up this data a little bit, it is always good measures to check for any missing values and find some way to handle this 

```python
#Checking for any missing data
state_data.isnull().sum()
district_data.isnull().sum()

# Handle missing values (if any)
state_data = state_data.dropna()  
district_data = district_data.dropna() 



```

## Step 2
Let's plot some data!


```python
#Making a bar graph of the states data
state_data.groupby('state')['homeschool_students'].sum().plot(kind='bar', figsize=(12, 6), color='skyblue')
plt.title('Homeschool Students Across 33 States')
plt.xlabel('State')
plt.ylabel('Total Homeschool Students')
plt.show()
```

This is a nice, from this graph we can easily see the state with the most homseschooled students, but it is a little harder to tell which states has the least amount of total students. We can fix this with a little bit more manipulation and with the use of python amazing librbaries such as seaborn 

```python
# Making a more detailed bar graph of the states data
plt.figure(figsize=(14, 8))

# Use seaborn for enhanced styling
sb.barplot(x='state', y='homeschool_students', data=state_data, errorbar=None, palette='viridis')
plt.title('Homeschool Students Across 33 States', fontsize=16)
plt.xlabel('State', fontsize=14)
plt.ylabel('Total Homeschool Students', fontsize=14)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right', fontsize=12)

for p in plt.gca().patches:#loop through each bar
    plt.gca().annotate(f"{int(p.get_height()):,}", (p.get_x() + p.get_width() / 2., p.get_height()),#adds commas and center the number
                       ha='center', va='center', xytext=(2.65, 10), textcoords='offset points', fontsize=9)
plt.show()
```

This updated bar graph enhances readability by providing a clearer representation of the x-axis. The exact number of homeschooled students for each state has been included, offering more detailed insights. Additionally, vibrant colors have been incorporated into the graph for enhanced visual appeal. With these modifications, we can now see that Florida leads among all 33 states, boasting the highest number of homeschooled students at 123,837, while the District of Columbia records the lowest at 705.

These findings are intriguing, gradually providing us with more data related to the previous questions we asked. Now, let's dive deeper into this dataset for further exploration.

## Note:

So far, we have graphed only general data, specifically the count of homeschooled students in each year. Let's attempt to visualize the trends across years and compare the data. This approach might provide valuable insights into addressing the questions we mentioned earlier.

```python
# Read data from CSV file again into a new variable
state_data_years= pd.read_csv("home_school_state.csv")

# extracting the invidual years
state_data_years['individual_year'] = state_data_years['year'].str.extract(r'(\d{4})').astype(int) #we use a regular expression to extract and then convert into a integer

# Group by the individual year, into a new variable
grouped_by_year = state_data_years.groupby('individual_year')

# Iterate through the years 2017 to 2023
for year in range(2017, 2023):
    # Access data for the current year
    states_current_year = grouped_by_year.get_group(year)
    
    # Display the result for the current year
    print(f"\nStates Data for the Year {year}:")
    display(states_current_year)
```

Great!, we have printed each invidual year for each state. Now, just like before I think it's important to plot this data so we can visually see the differences


## Plotting Each Year

```python
#extracting the years
state_data['individual_year'] = state_data['year'].str.extract(r'(\d{4})').astype(int)

#grouping them
grouped_by_year = state_data.groupby('individual_year')

# Plotting a bar graph for each year
for year in range(2017, 2023):
    # Access data for the current year
    states_current_year = grouped_by_year.get_group(year)
    
    # Check for missing values
    missing_states = states_current_year[states_current_year['homeschool_students'].isnull()]['state'].tolist()

    # Drop states with missing values
    states_current_year = states_current_year.dropna(subset=['homeschool_students'])

    #Handling State drop case
    if missing_states:
        print(f"For the year {year}, the following states have missing values and are dropped: {missing_states}")

    # Create a bar graph with state names at a 45-degree angle
    plt.figure(figsize=(12, 6))
    plt.bar(states_current_year['state'], states_current_year['homeschool_students'])
    plt.title(f'Homeschool Students Across States for Year {year}')
    plt.xlabel('State')
    plt.ylabel('Homeschool Students')
    plt.xticks(rotation=45, ha='right')  # Rotate state names at a 45-degree angle
    plt.tight_layout()  # Adjust layout for better readability
    plt.show()
```

**Notice**
For some years, certain states were not displayed due to it having no value for that specific year. For each year, with a missing state have been printed out to the user and made aware of. 

## Invidual Plotting
So far, a bar graph have been plotted for all states with a total number of homeschooled students to the user. Next, we plotted a bar graph to compare each state for that specific year. Next, I am going to take the top 3 states with the highest amount of homeschooled student's and focus on that. 

```python
state_data['individual_year'] = state_data['year'].str.extract(r'(\d{4})').astype(int)

# Group by state and sum the total homeschool students for each state
total_students_by_state = state_data.groupby('state')['homeschool_students'].sum()

# Get the top 3 states with the most homeschooled students
top_states = total_students_by_state.nlargest(3).index

# Filter the data for the top 3 states
top_states_data = state_data[state_data['state'].isin(top_states)]

# Plotting line graphs for each state and year
plt.figure(figsize=(12, 6))
sb.lineplot(x='individual_year', y='homeschool_students', hue='state', data=top_states_data)
plt.title('Homeschool Students Trend for Top 3 States')
plt.xlabel('Year')
plt.ylabel('Homeschool Students')
plt.legend(title='State')
plt.show()
```
